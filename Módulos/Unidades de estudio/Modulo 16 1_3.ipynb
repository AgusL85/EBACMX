### Añadiendo una columna nueva

```python
# Se hará el mismo tratamiento que se hizo en lecciones anteriores, para
# generar la duración de cada registro
# Como primer paso, se añadirá un nuevo campo de duración, para no modificar la estructura original
df["duration_num"] = df["duration"]
```

```python
# Además se generará un campo adicional de unidades, para guardar la información, que podrá ser "min" o "season"
df["duration_unit"]= " "
```

```python
# como primer paso, se cambia el tipo de la columna de las columnas de duration
df["duration_unit0"] = df["duration_unit"].astype(pd.StringDtype())
```

```python
df.dtypes
```

```python
# Si lo quisiera hacer para más columnas a la vez (todas las de tipo object)
# Podría primero buscar todas las columnas de ese tipo
df.select_dtypes(include="object").columns

# y luego aplicarlo
```

```python
# cambia el tipo de variable de todas las columnas tipo object
stringcols = df.select_dtypes(include="object").columns
df[stringcols] = df[stringcols].fillna("").astype(pd.StringDtype())
```

```python
df.dtypes
```

```python
df.head(3)
```

### Búsqueda de datos
* Uso de where, endswith, replace

```python
# luego, se buscará los registros que tengan "min", y con ellos, se asignará valores
# a la columna duration_unit

# el primer paso es definir a todas las unidades como "min", luego esto se cambiará
df["duration_unit"] = "min"
```

```python
df.head(3)
```

```python
# usando where asignamos la duration_unit a min cuando es "min" y a "season" cuando es <> min
df["duration_unit"] = df ["duration_unit"].where(df.duration_num.str.endswith("min"),"season")
```

```python
df.sample(5)
```

```python
# Se usa el método replace para el reemplazo de los strings
# En este caso se reemplaza todas las ocurrencias de min, Season y Seasons por blancos
df["duration_num"] = df["duration_num"].replace({"min":"","Season":"","Seasons":"","s":""},regex=True)
```

```python
df.sample(5)
```

```python
# Se valida el contenido y el split que se hizo de la columna
df[["show_id", "type", "duration" , "duration_num" , "duration_unit"]].sample(100)

#* Nota, se podía haber utilizado .split también, pero se quería recorrer algunos comando más
```

```python
# todavía tiene números y hay que convertir
df.dtypes
```

```python
# Se cambia el tipo de columna de duration_num a numérico
df["duration_num"] = pd.to_numeric(df["duration_num"])
```

```python
# Una vez con la columna en numérico, se la visualiza en el describe () !!
df.describe()
```

```python
# como no coincide la cantidad de registros entre una variable y otra, lo más seguro es que haya registros Nulos (null)
df.loc[df["duration_num"].isnull()]
```

```python
# en la importación, para 3 películas, se asignó la duración al rating
# También podría verlos por ID: df.loc[df["show_id"] == "s5542"]
# con la función at, puedo hacer los cambios de una manera 100% manual
# Si bien no es escalable a grandes cantidades de registros, es muy útil para cambios pequeños

df.at[5541,"duration_unit"] = "min"
df.at[5794,"duration_unit"] = "min"
df.at[5813,"duration_unit"] = "min"

df.at[5541,"duration_num"] = 74
df.at[5794,"duration_num"] = 84
df.at[5813,"duration_num"] = 66

df.at[5541,"duration"] = "74 min"
df.at[5794,"duration"] = "84 min"
df.at[5813,"duration"] = "66 min"
```

## Insertando, Eliminando, Edición de un DataFrame nuevo

```python
# nuevo dataframe

clase= [("Jorge", 22 , "MX"),
         ("Anita", 31 , "MX"),
         ("Raúl", 16 , "BR"),
         ("Simón", 41 , "BR"),
         ("Diego", 33 , "PA"),
         ("Jorge", 35 , "PA"),
         ("Almendra", 35 , "COL"),
         ("Mary", 35 , "COL"),
         ("Peter", 35 , "EC"),
         ("Andrés", 35 , "EC"),]

df_clase = pd.DataFrame(clase, columns = ["Nombre" , "Edad" , "País"])
```

```python
type(clase)
```

```python
df_clase
```

```python
# ordeno la clase por edad
df_clase.sort_values(by = "Edad", ascending =False)
```

```python
# llega un alumno nuevo
df_clase = df_clase.append({"Nombre" : "Isabel" , "Edad": 35, "País": "PER"}, ignore_index=True)
```

```python
df_clase
```

```python
# Para eliminar un registro
# Basado en el índice
df_clase.drop(df_clase.index[10])
```

```python
df_clase
```

```python
# por qué no se eliminó? Muy importante asigna el df a la función
df_clase = df_clase.drop(df_clase.index[10])
```

```python
df_clase
```

```python
# Podría eliminar más de un registro
# Tener cuidado el doble corchete!
df_clase.drop(df_clase.index[[8,9]])
```

```python
# Para modificar un registro
# se puede localizar con .at
df_clase.at[3, "Nombre"] = "José"
```

```python
df_clase
```

```python
# para más de un registro se puede usar .loc, y además hacerlo para múltiples columnas
df_clase.loc[1:2,["Edad", "País"]] = [32,"ARG"]
```

```python
df_clase
```

```python
# finalmente, se puede usar .replace
df_clase.replace ("PA" , "PAN", inplace= True)
```
