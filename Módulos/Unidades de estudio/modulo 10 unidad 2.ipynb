```python
import pandas as pd
df = pd.read_csv("StudentsPerformance.csv")
df
```

```python
import seaborn as sns
```

```python
for style in ["dark" , "whitegrid" , "darkgrid" , "ticks"]:
    sns.set_style(style)
    sns.displot(df["reading score"])
```

```python
# remociÃ³n de ejes
sns.set_style("ticks")
sns.displot(df["reading score"], kde = True)
sns.despine(left= True)
```

```python
# remociÃ³n de ejes
sns.set_style("ticks")
sns.displot(df["reading score"], kde = True)
sns.despine(left= True, bottom = True)
```

```python
# CombinaciÃ³n de parÃ¡metros con Matplotlib
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
sns.distplot(df["writing score"], ax = ax)
ax.set(xlabel= "Score de escritura", ylabel = "Frecuencia relativa")
```

```python
# Truncamiento de escala de visualizaciÃ³n
fig , ax = plt.subplots()
sns.distplot(df["writing score"], ax = ax)
ax.set(xlabel= "Score de escritura", ylabel = "Frecuencia relativa" , xlim = (0,90),
       title = "Histograma de frecuencias para puntajes de escritura")
```

```python
fig,(ax0, ax1) = plt.subplots(nrows = 1, ncols = 2 , sharey = True)
sns.distplot(df["writing score"] , ax = ax0) # Histogramas para TODAS las observaciones
sns.distplot(df.query ('gender == "male"' ) ["writing score"] , ax = ax1) #histograma para los hombres
ax0.set(xlabel = "Puntaje general de escritura")
ax1.set(xlabel = "Puntaje de escritura para hombres")
ax1.axvline(x= 75,label="Objetivo" , linestyle = "--")
ax.legend()
```

```python
# asignaciÃ³n de colores
paleta = ["deep", "muted" , "pastel" , "bright" , "dark" , "colorblind"]

for p in paleta: 
    sns.set_palette(p)
    sns.displot(df["reading score"] , kde = True)
```
