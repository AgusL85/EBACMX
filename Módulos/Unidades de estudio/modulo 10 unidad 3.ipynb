```python
import pandas as pd
df = pd.read_csv("StudentsPerformance.csv")
df
```

```python
# CreaciÃ³n de un stripplot (grÃ¡fico de franjas/Bandas)

import seaborn as sns
sns.stripplot(data = df , y = "writing score" , x= "math score" , jitter = True)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
sns.stripplot(data = df , y = "writing score" , x= "math score" , jitter = True)
ax.axes.xaxis.set_ticks([])
plt.show
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
sns.stripplot(data = df , y = "writing score" , x= "race/ethnicity" , jitter = True)
plt.show
```

```python
# CreaciÃ³n de un swarmplot (Diagrama de enjambre)
# Busca colocar las observaciones de manera que no se empalmen 

sns.swarmplot(data = df , y = "writing score" , x= "race/ethnicity")
```

```python
# CreaciÃ³n de un boxplot (diagrama de caja)
sns.boxplot(data = df , y = "writing score" , x= "race/ethnicity")
```

```python
# CreaciÃ³n de grÃ¡fico de violÃ­n 
#mezcla de un boxplot con funciÃ³n de densidad Kernel
sns.violinplot(data = df , y = "writing score" , x= "race/ethnicity")
```

```python
# CreaciÃ³n de un lvplot (letter-value-plot)
# Mezcla de boxplot y el violinplot, conveniente para bases de datos grandes
sns.catplot(x = "race/ethnicity" , y ="writing score" , kind="boxen", data = df)
```

```python
# CreaciÃ³n de un barplot
sns.barplot(data = df, y = "writing score" , x= "race/ethnicity")
```

```python
# CreaciÃ³n de un barplot (grÃ¡fico de barras)
sns.barplot(data = df, y = "writing score" , x= "race/ethnicity", hue= "lunch")
```

```python
# CreaciÃ³n de un pointplot
sns.pointplot(data = df, y = "writing score" , x= "race/ethnicity", hue= "lunch")
```

```python
# creaciÃ³n de un countplot (grÃ¡fico de puntos)
sns.countplot(data = df, y= "race/ethnicity", hue= "lunch")
```

```python
# GrÃ¡ficos de regresiÃ³n con Regplot
sns.regplot( data = df, y = "writing score" , x= "reading score" , marker= ".")
```

```python
# GrÃ¡fico de residuales
sns.residplot( data = df, y = "writing score" , x= "reading score" )
```

```python
# GrÃ¡ficos de regresiÃ³n con regplot y media por valor de x
# Muy Ãºtil para enfatizar tendencias
import numpy as np
sns.regplot( data = df, y = "writing score" , x= "reading score" , x_estimator = np.mean)
```

```python
# CreaciÃ³n de un mapa de calor para una matriz de correlaciÃ³n 
sns.heatmap(df.corr())
```

```python
sns.heatmap(df.corr(), annot= True)
```

```python
# CreaciÃ³n de una tabla cruzada

pd.crosstab(df["gender"] , df["race/ethnicity"], values = df["math score"] , aggfunc = "mean")
```

```python
# CreaciÃ³n de un mapa de calor para una tabla cruzada
sns.heatmap(pd.crosstab(df["gender"] , df["race/ethnicity"], values = df["math score"] , aggfunc = "mean"), annot= True)
```
