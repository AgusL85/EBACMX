```python
# Ejemplos con break
while True:
    linea=input ("Escribe un texto: ")
    if linea=="Fin":
        break
    print(linea)
print ("Proceso terminado")
```

```python
# Ejemplos con continue
while True:
    linea=input ("Escribe un texto: ")
    if linea[0]=="#":
        continue
    if linea=="Fin":
        break
    print (linea)
print ("Proceso terminado")
```

```python
#Ejercicio 2: 
#Considere el conjunto de números: 1, 2, ..., 9 Desarrolle un programa en Python que determine los valores de la lista 
#anterior que son pares así como los impares

numeros=(1,2,3,4,5,6,7,8,9) #creamos una tupla
valores=len(numeros)
impares=0
pares=0

contador=1
while contador <= valores:
    if numeros[contador - 1] % 2 == 0:
        pares += 1
    else:
        impares += 1
    contador += 1
    
print("Número valores pares: ",pares)
print("Número valores impares: ",impares)
    
```

```python
# Ejercicio 3
# La secuencia de números Fibonacci está dada de la siguiente manera: 0, 1, 1, 2, 3, 5, ...
# Desarrolle un programa en Python que encuentre los números de esta secuencia entre 0 y 50

x,y = 0,1
print(x)
print(y)

while x+y < 50:
    x = x + y
    print(x) #se recaluclo la x en el renglón anterior
    y = x + y
    print (y) #se recalcula la y en el renglón anterior 
```
