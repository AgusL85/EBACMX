```python
import pandas as pd
df = pd.read_csv("StudentsPerformance.csv")
df
```

```python
# histograma de pandas
df ["math score"].plot.hist()
```

```python
# Histograma en Seaborn

import seaborn as sns
sns.distplot(df["math score"])
```

```python
# histograma alternatuvo en Seaborn
sns.displot(df["math score"])
```

```python
sns.displot(df["math score"], kde = True, bins=10)
```

```python
sns.distplot(df["math score"] , hist = False)
```

```python
sns.distplot(df["math score"] , hist = False, kde_kws = {"shade" : True})
```

```python
# CreaciÃ³n de diagramas de dispersiÃ³n con recta de dispersiÃ³n
sns.regplot (x= "reading score" , y= "writing score" , data= df)
```

```python
sns.lmplot (x= "reading score" , y= "writing score" , data= df)
```

```python
# GrÃ¡fico de dispersiÃ³n con regresiÃ³n lineal por categorÃ­a de gÃ©nero
sns.lmplot (x= "reading score" , y= "writing score" , data= df, hue = "gender")
```
