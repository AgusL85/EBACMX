```python
import numpy as np
import pandas as pd
```

```python
arreglo = np.array([["Bella","Labrador","Cafe",56,24,"2013-07-01"],["Charlie","Poodle","Negro",43,24,"2016-09-16"],
                    ["Lucy","Chow Chow","Café",46,24,"2014-08-25"],["Cooper","Shnauzer","Gris",49,17,"2011-12-11"],
                    ["Bernie","San Bernardo","Blanco",77,74,"2018-02-27"],["Max","Chow Chow","Blanco",45,28,"2019-03-27"],
                    ["Reed","Shnauzer","Negro",45,15,"2017-09-18"] ])
arreglo
```

```python
df= pd.DataFrame(data=arreglo, columns= ["Nombre", "Raza", "Color","Altura_cm","Peso_kg", "Fecha_nac"])
df["Altura_cm"] = df ["Altura_cm"].astype(int) #para convertir los caracteres a números enteros
df["Peso_kg"] = df ["Peso_kg"].astype(int)
df
```

```python
# ejercicio 2

# Mediante el uso de una tabla pivote sobre el DataFrame de mascotas, extraiga los registros donde la raza de la mascota es
# “Labrador”, “Poodle” y “San Bernardo”

#Creamos la lista pivote
df_altura_por_raza_color = df.pivot_table("Altura_cm" , index = "Raza" , columns ="Color")
df_altura_por_raza_color
```

```python
# Extraemos los registros solicitados

df_altura_por_raza_color.loc  ["Labrador" : "San Bernardo"]
```

```python
# Cálculo de promedios para TODOS los renglones (por columna)
df_altura_por_raza_color.mean (axis="index")
```

```python
# Cálculo de promedios para TODOS las columnas (por renlgón)
df_altura_por_raza_color.mean (axis="columns")
```
